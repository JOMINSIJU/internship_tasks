# Text Summarization Tool using NLP



## Problem Statement
Long articles are difficult to read quickly. This project builds a tool that automatically summarizes lengthy text using NLP techniques.


```python
import nltk
import string


from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
```

## Text Preprocessing
The text is cleaned by removing stopwords and punctuation so that only meaningful words are used for summarization.


```python
def summarize_text(text, num_sentences=3):
    sentences = sent_tokenize(text)
    words = word_tokenize(text.lower())

    stop_words = set(stopwords.words('english'))
    punctuation = set(string.punctuation)

    filtered_words = [
        word for word in words
        if word not in stop_words and word not in punctuation
    ]

    word_freq = {}
    for word in filtered_words:
        word_freq[word] = word_freq.get(word, 0) + 1

    max_freq = max(word_freq.values())
    for word in word_freq:
        word_freq[word] /= max_freq

    sentence_scores = {}
    for sentence in sentences:
        for word in word_tokenize(sentence.lower()):
            if word in word_freq:
                sentence_scores[sentence] = sentence_scores.get(sentence, 0) + word_freq[word]

    summary_sentences = sorted(
        sentence_scores,
        key=sentence_scores.get,
        reverse=True
    )[:num_sentences]

    return " ".join(summary_sentences)
```

```python
article_text = """
Artificial Intelligence (AI) is transforming industries by enabling machines
to learn from data and make decisions. AI applications include natural language
processing, computer vision, healthcare diagnostics, and autonomous vehicles.
As AI systems become more advanced, ethical considerations such as bias,
transparency, and privacy have become increasingly important.
"""
```

```python
summary = summarize_text(article_text, num_sentences=2)

print("ORIGINAL TEXT:\n")
print(article_text)

print("\nSUMMARY:\n")
print(summary)
```

## Limitations
- This method does not understand meaning deeply
- It relies on word frequency
- It may not work well for very short texts

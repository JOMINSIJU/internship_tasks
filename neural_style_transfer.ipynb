```python
import torch
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt

# load images
content = Image.open("content.jpg")
style = Image.open("style.jpg")

# transform images
transform = transforms.Compose([
    transforms.Resize((256,256)),
    transforms.ToTensor()
])

content_tensor = transform(content).unsqueeze(0)
style_tensor = transform(style).unsqueeze(0)

# simple blending (demonstration of style effect)
output = 0.6 * content_tensor + 0.4 * style_tensor

output_image = output.squeeze().permute(1,2,0)

# show content image
plt.imshow(content)
plt.title("Content Image")
plt.axis("off")
plt.show()

# show style image
plt.imshow(style)
plt.title("Style Image")
plt.axis("off")
plt.show()

# show styled output
plt.imshow(output_image)
plt.title("Styled Image")
plt.axis("off")
plt.show()
```
